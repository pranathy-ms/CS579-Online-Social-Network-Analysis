```python
import networkx as nx
import matplotlib.pyplot as plt

#       Size Avg. degree
#E.Coli 282 7.35 
#C.Elegans 282 14 

# Given parameters
n = 282  # Number of nodes
avg_degree = 7.35  # Desired average degree

# Calculate probability parameter p
p = avg_degree / (n - 1)

print("Probability parameter (p):", p)

# Generate a random graph 
random_graph_model = nx.erdos_renyi_graph(n, p)
if nx.is_connected(random_graph_model):
        avg_path_length = nx.average_shortest_path_length(random_graph_model)
else:
        largest_cc = max(nx.connected_components(random_graph_model), key=len)
        largest_cc_graph = random_graph_model.subgraph(largest_cc)
        avg_path_length = nx.average_shortest_path_length(largest_cc_graph)

print("Average Path Length:", avg_path_length)

# Calculate average clustering coefficient
avg_clustering_coefficient = nx.average_clustering(random_graph_model)
print("Average Clustering Coefficient:", avg_clustering_coefficient)

# Visualize the graph
plt.figure(figsize=(10,6))
pos = nx.spring_layout(random_graph_model)
nx.draw(random_graph_model, pos, node_size=50, with_labels=False, edge_color='gray',alpha=0.6)
plt.show()
```
